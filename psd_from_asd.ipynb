```python
import numpy as np

# あなたのASDファイルのパス
input_files = [
    '../asd_files/aLIGO_O4_high_asd.txt',
    '../asd_files/AdV_asd.txt'
]

for f in input_files:
    # データを読み込む
    data = np.loadtxt(f)
    freq = data[:, 0]
    asd = data[:, 1]
    
    # 二乗してPSDにする
    psd = asd ** 2
    
    # 新しいファイル名で保存 (例: _psd.txt をつける)
    new_filename = f.replace('_asd.txt', '_psd.txt')
    np.savetxt(new_filename, np.column_stack((freq, psd)))
    print(f"Created: {new_filename}")
```
