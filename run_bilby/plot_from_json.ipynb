```python
import configparser
import argparse
import os
from single_posterior_plot import plot_posterior
import errno

"""set up"""
#recieve the argments from command line
def plot_from_json(config_file_path):

    config_path = config_file_path
    config_ini = configparser.ConfigParser()
    config_ini.read(config_path)

    if not os.path.exists(config_path):
        raise FileNotFoundError(errno.ENOENT, os.strerror(errno.ENOENT), config_path)
    
    config_setting = config_ini['setting']

    """set the event to be analyzed"""
    event_name = config_setting['event_name']
    #index = config_setting['index']
    #index = 'v1'
    index = None
    if index is not None:
        outdir = 'outdirs/outdir_' + event_name + '_' + index
        label = event_name + '_' + index
    else:
        outdir = 'outdirs/outdir_' + event_name
        label = event_name

    plot_posterior(
                    path_json = outdir+'/'+label+'_result.json',
                    path_outdir = outdir,
                    default_plot_parameters = True,
                    show_fig = False,
                    plot_kde=False,
                    )
```

```python
file_path_list = [
                'shiftIm_to_220_dw0.1w1_snr100_EPparam_Heaviside.ini',   
                'shiftIm_to_220_dw0.01w1_snr100_EPparam_Heaviside.ini',   
                'shiftIm_to_220_dw0.001w1_snr100_EPparam_Heaviside.ini',
                'shiftRe_to_220_dw0.1w1_snr100_EPparam_Heaviside.ini',   
                'shiftRe_to_220_dw0.01w1_snr100_EPparam_Heaviside.ini',   
                'shiftRe_to_220_dw0.001w1_snr100_EPparam_Heaviside.ini',   
]

for file_path in file_path_list:
    config_file_path= f'./config/{file_path}'
    plot_from_json(config_file_path)
```
