## $$\delta\omega = \omega_1 - \omega_2 = 2\pi(f_1-f_2) + i(\frac{1}{\tau_1}-\frac{1}{\tau_2})$$
## $$f_2 = f_1 - \delta\omega/2\pi = f_1 - \frac{|\omega_1|}{2\pi}|\frac{\delta\omega}{\omega_1}| \qquad \mathrm{if} \quad \tau_1=\tau_2 \quad (\mathrm{and} \quad f_1>f_2)$$
## $$\tau_2 = \frac{1}{\frac{1}{\tau_1}+\delta\omega} = \frac{1}{\frac{1}{\tau_1}+|\omega_1||\frac{\delta\omega}{\omega_1}|} \qquad \mathrm{if} \quad f_1=f_2 \quad (\mathrm{and} \quad \tau_1>\tau_2)$$

```python
import numpy as np
import astropy.constants as const

###constants###
Mo = const.M_sun.value #solar mass [kg]
G = const.G.value #Newton constant [m^3 kg^-1 s^-2]
c = const.c.value #light speed [m s^-1]
pc = const.pc.value #1pc [m]
###############

M = 60 * Mo
Rs = 2 * G * M / c**2 #Schwarzschild radius [m]

alpha = G * M / c**3
print(f'alpha = {alpha} s')
print()

f1 = 201.23779332666524 #a/M=0, 220, M=60
tau1 = 0.0033219623034593474 #a/M=0, 220, M=60
w1 = 2 * np.pi * f1 + 1j / tau1

# dw_over_w1 = 0.1
dw_over_w1 = 0.3
print(f'f1={f1}')
print(f'tau1={tau1}')
print(f'|dw| = {dw_over_w1 * np.abs(w1)}')
print(f'1/|dw| = {-1 / (dw_over_w1 * np.abs(w1))}')
print()

"""real part shift"""
print('\nreal part shift')
tau2 = tau1
f2 = f1 - (dw_over_w1 * np.abs(w1)) / (2 * np.pi)
dw = 2 * np.pi * (f1 - f2)
print(f"f2 = {f2}")
print(f'dw = {dw}')
print(f'|dw/w1| = {np.abs(dw / w1)}')

"""imaginary part shift"""
print('\nimaginary part shift')
f2 = f1
tau2 = 1 / (1/tau1 + (dw_over_w1 * np.abs(w1)))
dw = 1j * (1/tau1 - 1/tau2)
print(f'tau2 = {tau2}')
print(f'dw = {dw}')
print(f'|dw/w1| = {np.abs(dw / w1)}')

alpha_dw = alpha * dw
print(f'\nalpha * dw = {alpha_dw}')
```

## $$f_2 = f_1 - \frac{|\delta\omega|}{2\pi}\cos\theta$$
## $$\tau_2 = \frac{1}{\frac{1}{\tau_1} - |\delta\omega|\sin\theta}$$

```python
import numpy as np
import astropy.constants as const

###constants###
Mo = const.M_sun.value #solar mass [kg]
G = const.G.value #Newton constant [m^3 kg^-1 s^-2]
c = const.c.value #light speed [m s^-1]
pc = const.pc.value #1pc [m]
###############

M = 60 * Mo
Rs = 2 * G * M / c**2 #Schwarzschild radius [m]

alpha = G * M / c**3
print(f'alpha = {alpha} s')
print()

f1 = 201.23779332666524 #a/M=0, 220, M=60
tau1 = 0.0033219623034593474 #a/M=0, 220, M=60
w1 = 2 * np.pi * f1 + 1j / tau1

dw_over_w1 = 0.001
# dw_over_w1 = 0.3

theta = 1 * np.pi/4

print(f'f1 = {f1}')
print(f'tau1 = {tau1}')
print(f'|dw| = {dw_over_w1 * np.abs(w1)}')
print(f'1/|dw| = {-1 / (dw_over_w1 * np.abs(w1))}')

"""shift both real and imaginary part"""
print('\nshift both real and imaginary part')
f2 = f1 - (dw_over_w1 * np.abs(w1)) / (2 * np.pi) * np.cos(theta)
tau2 = 1 / (1/tau1 + (dw_over_w1 * np.abs(w1)) * np.sin(theta))
dw = 2 * np.pi * (f1 - f2) + 1j * (1/tau1 - 1/tau2)
print(f"f2 = {f2}")
print(f'tau2 = {tau2}')
print(f'dw = {dw}')
print(f'|dw/w1| = {np.abs(dw / w1)}')

alpha_dw = alpha * dw
print(f'\nalpha * dw = {alpha_dw}')
```
