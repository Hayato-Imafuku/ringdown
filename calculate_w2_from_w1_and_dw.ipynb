```python
import numpy as np
import astropy.constants as const

###constants###
Mo = const.M_sun.value #solar mass [kg]
G = const.G.value #Newton constant [m^3 kg^-1 s^-2]
c = const.c.value #light speed [m s^-1]
pc = const.pc.value #1pc [m]
###############

M = 60 * Mo
Rs = 2 * G * M / c**2 #Schwarzschild radius [m]

omega_225 = 1.0738334 - 1j * 1.4974469
omega_226 = 1.2044549 - 1j * 1.5406739
delta_omega = omega_226 - omega_225
print(f"|delta_omega|/|omega_225| = {np.abs(delta_omega) / np.abs(omega_225)}")
print()

omega_225_dimensional = omega_225 * (c/Rs)
omega_226_dimensional = omega_226 * (c/Rs)
print(f"omega_225_dimensional = {omega_225_dimensional}")
print(f"omega_226_dimensional = {omega_226_dimensional}")
print()
print(f"f_225 = {omega_225_dimensional.real / (2 * np.pi)}")
print(f"f_226 = {omega_226_dimensional.real / (2 * np.pi)}")
print(f"tau_225 = {1 / omega_225_dimensional.imag}")
print(f"tau_226 = {1 / omega_226_dimensional.imag}")
print()

dw = omega_226_dimensional - omega_225_dimensional
abs_dw = np.abs(dw)
print(f"|dw| = {abs_dw}")
print()

dw_over_w225 = abs_dw / np.abs(omega_225_dimensional)
print(f'|dw|/|w_225| = {dw_over_w225}')
```
