```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const

###constants###
Mo = const.M_sun.value #solar mass [kg]
G = const.G.value #Newton constant [m^3 kg^-1 s^-2]
c = const.c.value #light speed [m s^-1]
pc = const.pc.value #1pc [m]
###############

"""calculate fundamental QNM frequency for a Schwarzschild black hole"""
"""maggiore Vol.2 P160"""

# print(const.M_sun)
# print(const.G)
# print(const.c)
# print(const.pc)

M = 60 * Mo
Rs = 2 * G * M / c**2 #Schwarzschild radius [m]

alpha = G * M / c**3
print(f'alpha = {alpha} s')

f = 0.747 * c / (2*np.pi) / Rs
# f = 12 * 1e3 * Mo / M

tau = Rs / (0.178 * c)

print(f'M = {M}, f = {f} Hz, tau = {tau} s')
print()
print(f'w_real = {2 * np.pi * f}, w_imag = {-1 / tau}')
print()

w1_real = 0.747343 * c / Rs
f1 = w1_real / (2 * np.pi)
tau1 = Rs / (-0.177925 * c)
print(f'w1_real = {w1_real} Hz')
print(f'f1 = {f1} Hz, tau1 = {tau1} s')
print()
w2_real = 0.693422 * c / (2 * Rs)
f2 = w2_real / (2 * np.pi)
tau2 = Rs / (-0.547830 * c)
print(f'w2_real = {w2_real} Hz')
print(f'f2 = {f2} Hz, tau2 = {tau2} s')
print()

w1 = w1_real + 1j / tau1
w2 = w2_real + 1j / tau2

delta_w = w2 - w1
alpha_delta_w = alpha * delta_w
print(f'delta_w = {delta_w}, alpha * delta_w = {alpha_delta_w}')
```
